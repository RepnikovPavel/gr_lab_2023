```python
from tslearn import generators
import numpy as np
import matplotlib.pyplot as plt
```

```python
N  = 1000
x_= np.arange(0,N)
y_=  generators.random_walks(n_ts=1,sz=N).flatten()
a_ = np.min(y_)
b_ = np.max(y_)
y_ = (y_-a_)/(b_-a_)
c_ = 60
d_ = 120
y_ *= (d_-c_)
y_ += c_
```
